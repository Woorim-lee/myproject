```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.family'] = "Malgun Gothic"
#폰트 바꾸기
```

```python
pizza = pd.read_csv('./pizza_09.csv')
pizza
```

```python
chicken07 = pd.read_csv('./chicken_07.csv')
chicken08 = pd.read_csv('./chicken_08.csv')
chicken09 = pd.read_csv('./chicken_09.csv')
chicken = pd.concat([chicken07, chicken08, chicken09])
chicken = chicken.reset_index(drop=True)
chicken
```

```python
pizza_sum = pizza.groupby('요일')['통화건수'].sum()
chicken_sum = chicken.groupby('요일')['통화건수'].sum()

weeks = ["월", "화", "수", "목", "금", "토", "일"]

pizza_sorted = pizza_sum.reindex(weeks)
chicken_sorted = chicken_sum.reindex(weeks)
```

```python
pizza_sorted
```

```python
chicken_sorted
```

```python
plt.figure(figsize=(8,5))
plt.title('요일에 따른 피자,치킨 주문량')
plt.bar(chicken_sorted.index, chicken_sorted)
plt.bar(pizza_sorted.index, pizza_sorted)
plt.show()
```
