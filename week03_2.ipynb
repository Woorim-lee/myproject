```python
import numpy as np
from PIL import Image
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
text = open('./data/Sequence_01.txt', 'rt', encoding='utf8')
text = text.read()
text = text.replace('\n', " ")
text

text2 = open('./data/Sequence_02.txt', 'rt', encoding='utf8')
text2 = text2.read()
text2 = text2.replace('\n', " ")
text2
```

```python
result = ""
for number in range(1,15):
    index = '{:02}'.format(number)
    filename = "Sequence_" + index + ".txt"
    print(filename)
    text = open('./data/' + filename, 'r', encoding='utf-8-sig')
    result += text.read().replace("\n", " ")

result
```

```python
import re
pattern = '[^\w\s]'
text = re.sub(pattern=pattern, repl='', string=result)
text
```

```python
import matplotlib.font_manager as fm

for f in fm.fontManager.ttflist:
    if 'Gothic' in f.name:
        print(f.fname)
```

```python
font_path = 'C:\Windows\Fonts\malgun.ttf'

wc = WordCloud(font_path=font_path, background_color='white')
wc.generate(text)

plt.figure(figsize=(50,50))
plt.axis('off')
plt.imshow(wc)
plt.show()
```

```python
# Generate a word cloud image
mask = np.array(Image.open('./data/sparta.png'))
wc = WordCloud(font_path=font_path, background_color="white", mask=mask)
wc.generate(text)

f = plt.figure(figsize=(50,50))
f.add_subplot(1,2, 1)
plt.imshow(mask, cmap=plt.cm.gray)
plt.title('Original Stencil', size=40)
plt.axis("off")
f.add_subplot(1,2, 2)
plt.imshow(wc, interpolation='bilinear')
plt.title('Sparta Cloud', size=40)
plt.axis("off")
plt.show()
```

```python
mask = np.array(Image.open('./data/cartoon-rock-musician.jpg'))
wc = WordCloud(font_path=font_path, background_color="white", mask=mask)
wc.generate(text)

f = plt.figure(figsize=(50,50))
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.imshow(wc, interpolation='bilinear')
plt.title('나의 워드클라우드', size=40)
plt.axis('off')
plt.show()
f.savefig('./data/myWordCloud.png')
```
